```python
data = pd.read_csv('commonsense/cm_test_hard_100.csv')
```

```python
import pandas as pd

arr = data["input"].tolist()

# Example: your DataFrame
df = pd.DataFrame({
    "prompt": arr,
})

# Convert each row to the required message format
def to_finetune_format(row):
    return {
        "messages": [
            {"role": "user", "content": row["prompt"]},
            {"role": "assistant", "content": "Your response should be either 0 or 1, where 0 means the action is commonsense and 1 means the action is not commonsense."},
        ]
    }

formatted = df.apply(to_finetune_format, axis=1).tolist()
print(formatted)   
```

```python
import json

with open("formatted_data.jsonl", "w") as f:
    for entry in formatted:
        f.write(json.dumps(entry) + "\n")
```
