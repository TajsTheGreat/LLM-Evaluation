```python
import matplotlib.pyplot as plt
```

```python

# Categories (x-axis)
categories = ['Commonsense', 'Justice', 'Deontology', 'Utilitarianism', 'Virtue Ethics']

# Accuracy values for each model
model_a_acc = [0.94, 0.872, 0.782, 0.576, 0.766]
model_b_acc = [0.934, 0.646, 0.64, 0.688, 0.682]
model_c_acc = [0.97, 0.902, 0.828, 0.602, 0.854]

# Optional: Model names
models = ['Mistral', 'ChatGPT 4o-mini', 'Gemini 2.0 Flash']
```

```python
plt.figure(figsize=(10, 6))

# Plot each model
plt.plot(categories, model_a_acc, marker='o', label='Mistral')
plt.plot(categories, model_b_acc, marker='s', label='ChatGPT o4-mini')
plt.plot(categories, model_c_acc, marker='^', label='Gemini 2.0 Flash')

# Add labels and title
plt.title('LLM Performance on Ethical Reasoning Tasks')
plt.xlabel('Categories')
plt.ylabel('Accuracy')
plt.ylim(0.5, 1.0)
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
